```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([0,2,4,6,8,10,12,14]);
r1 = np.array([303.02, 300.88, 299.73, 299.26, 299.10, 298.96, 298.82, 298.67])
plt.plot(time, r1, 'o');
plt.axvline(x=0)

m, b = np.polyfit(time[4:], r1[4:], 1)
plt.plot(time, m * time + b )


plt.title('Reaction Profile of Trial 1 at 303K')
plt.xlabel('Reaction Progress (s) ± 0.001s')
plt.ylabel('Temperature (K) ± 0.01 K')

plt.show()
c = 303-b
print('Corrected Final Temperature of Trial 1: ', "{:.2f}".format(b), 'K') 
print('Temperature Change: ', "{:.2f}".format(c), 'K')
```
