```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([0,2,4,6,8,10,12,14]);
r1 = np.array([298.00, 295.87, 294.75, 294.26, 294.09, 293.93, 293.77, 293.62,])
plt.plot(time, r1, 'o');
plt.axvline(x=0)

m, b = np.polyfit(time[5:], r1[5:], 1)
plt.plot(time, m * time + b )


plt.title('Reaction Profile of Trial 1 at 298K')
plt.xlabel('Reaction Progress (s) ± 0.001s')
plt.ylabel('Temperature (K) ± 0.01 K')

plt.show()
c = 298-b
print('Corrected Final Temperature of Trial 1: ', "{:.2f}".format(b), 'K') 
print('Temperature Change: ', "{:.2f}".format(c), 'K')
```
