```python
import numpy as np
import matplotlib.pyplot as plt

time = np.array([]);
for x in range(0,21):
    time = np.append(time, x * 30)
r1 = np.array([20.0, 20.5, 20.9, 21.4, 22.0, 22.3, 22.7,23,23.2,23.3,23.3,23.3,23.2,23.1,23.1,23.0,22.9,22.8,22.6,22.6,22.4])
plt.plot(time, r1, 'o');
plt.axvline(x=0)

m, b = np.polyfit(time[12:], r1[12:], 1)
plt.plot(time, m * time + b )




plt.title('Reaction Profile of Reaction 1')
plt.xlabel('Reaction Time (s) ± 1s')
plt.ylabel('Temperature (°C)')

plt.show()
print('Corrected Final Temperature: ', "{:.1f}".format(b), '°C') 
```
